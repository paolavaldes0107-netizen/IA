```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("cleaned_dataset.csv")

# Ver las primeras filas
df.head()
```

```python
# Tomamos 40 datos al azar
datos = df.sample(n=40, random_state=1)

# 20 para entrenar y 20 para probar
train = datos.iloc[:20]
test = datos.iloc[20:]

X_train = train.drop("Outcome", axis=1)
y_train = train["Outcome"]

X_test = test.drop("Outcome", axis=1)
y_test = test["Outcome"]
```

```python
import math

def distancia_euclidiana(x, y):
    suma = 0
    for i in range(len(x)):
        suma += (x[i] - y[i]) ** 2
    return math.sqrt(suma)
```

```python
punto1 = [1,106,70,28,135,34.2,0.142,22]
punto2 = [2,102,86,36,120,45.5,0.127,23]

distancia = distancia_euclidiana(punto1, punto2)
print("Distancia euclidiana:", distancia)
```

```python
def knn_manual(X_train, y_train, punto_test, k=3):
    distancias = []

    for i in range(len(X_train)):
        d = distancia_euclidiana(
            X_train.iloc[i].values,
            punto_test.values
        )
        distancias.append((d, y_train.iloc[i]))

    # Ordenamos por distancia
    distancias.sort(key=lambda x: x[0])

    # Tomamos los k vecinos más cercanos
    vecinos = distancias[:k]

    # Votación
    clases = [v[1] for v in vecinos]
    return max(set(clases), key=clases.count)
```

```python
resultados = []

for i in range(len(X_test)):
    prediccion = knn_manual(X_train, y_train, X_test.iloc[i], k=3)
    resultados.append([y_test.iloc[i], prediccion])

tabla = pd.DataFrame(resultados, columns=["Resultado Real", "Resultado Predicho"])
tabla
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("Outcome", axis=1)
y = df["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=3, metric="euclidean")
knn.fit(X_train, y_train)

pred = knn.predict(X_test)
accuracy_raw = accuracy_score(y_test, pred)

accuracy_raw
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_mm = scaler.fit_transform(X_train)
X_test_mm = scaler.transform(X_test)

knn_mm = KNeighborsClassifier(n_neighbors=3)
knn_mm.fit(X_train_mm, y_train)

pred_mm = knn_mm.predict(X_test_mm)
accuracy_mm = accuracy_score(y_test, pred_mm)

accuracy_mm
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

knn_std = KNeighborsClassifier(n_neighbors=3)
knn_std.fit(X_train_std, y_train)

pred_std = knn_std.predict(X_test_std)
accuracy_std = accuracy_score(y_test, pred_std)

accuracy_std
```

```python
pd.DataFrame({
    "Metodo": ["Sin escalar", "Normalizado", "Estandarizado"],
    "Accuracy": [accuracy_raw, accuracy_mm, accuracy_std]
})
```
